```python
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
def set_style():
    plt.style.use(['seaborn-ticks', 'seaborn-notebook'])
    matplotlib.rc("font", family="Arial")
```

### RMSD

```python
df1 = pd.read_csv("trajrmsd2.dat", sep=r'\s+', skiprows=1,usecols=[0, 1], 
                  names=['frames', 'RMSD'])
df2 = pd.read_csv("trajrmsd2.dat", sep=r'\s+', skiprows=1,usecols=[0, 2], 
                  names=['frames', 'RMSD'])
df3 = pd.read_csv("trajrmsd2.dat", sep=r'\s+', skiprows=1,usecols=[0, 3], 
                  names=['frames', 'RMSD'])
```

```python
df1['Version']='Amber 16'
df2['Version']='Amber 18'
df3['Version']='Amber 18 on BP'
```

```python
df=pd.concat([df1,df2,df3])
```

```python
df["Time (ns)"]=(df["frames"]*50+50)/1000
df=df.rename(columns={"RMSD": r"RMSD ($\AA$)"})
df
```

```python
sns.relplot(x="Time (ns)", y=r"RMSD ($\AA$)", hue="Version",
            kind="line", data=df);
plt.savefig('./RMSD.png', bbox_inches='tight', dpi=300)
plt.show()
```

### Energy

```python
df1 = pd.read_csv("ETOT_1.dat", sep=r'\s+', skiprows=1,usecols=[0, 1], 
                  names=['frames', 'Etot'])
df2 = pd.read_csv("ETOT_2.dat", sep=r'\s+', skiprows=1,usecols=[0, 1], 
                  names=['frames', 'Etot'])
df3 = pd.read_csv("ETOT_6.dat", sep=r'\s+', skiprows=1,usecols=[0, 1], 
                  names=['frames', 'Etot'])
```

```python
df1['Etot'][0]
```

```python
df1['Version']='Amber 16'
df2['Version']='Amber 18'
df3['Version']='Amber 18 on BP'

df1['Etot_rel']=df1['Etot']-df1['Etot'][0]
df2['Etot_rel']=df2['Etot']-df2['Etot'][0]
df3['Etot_rel']=df3['Etot']-df3['Etot'][0]

df=pd.concat([df1,df2,df3])

df["Time (ns)"]=(df["frames"]*10)/1000
df=df.rename(columns={"Etot": r"$E_{tot}$ (kcal/mol)",'Etot_rel':"Relative Potential Energy (kcal/mol)" })
df
```

```python
sns.relplot(x="Time (ns)", y=r"$E_{tot}$ (kcal/mol)", hue="Version",
            kind="line", data=df);
plt.savefig('./E_tot.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
sns.set(style="ticks", rc={"lines.linewidth": 1})
sns.relplot(x="Time (ns)", y="Relative Potential Energy (kcal/mol)", hue="Version",
            kind="line", data=df);
plt.savefig('./E_tot_rel.png', bbox_inches='tight', dpi=400)
plt.show()
```
